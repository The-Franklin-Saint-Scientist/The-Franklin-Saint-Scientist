How many unique goods did the business sell overall?

```python
Number_of_items = df_1["PRODUCTLINE"].nunique()
print(f"The number of unique items sold are:",Number_of_items)
```

List all the standout items the business sells.

```python
List_of_items = df_1["PRODUCTLINE"].unique()
print(f"The list of items sold are = ,{List_of_items}".upper())
```

Average cost for each product.

```python
Product_Price = df_1.groupby("PRODUCTLINE") ["PRICEEACH"].mean()\
.reset_index()\
.sort_values("PRODUCTLINE")\
.reset_index(drop=True)
Product_Price.columns = ["Product Name", "Average price"]
Product_Price.index += 1
Product_Price
```

How many of each product did the business sell in total?

```python
Units_of_each_product = df_1.groupby("PRODUCTLINE") ["QUANTITYORDERED"].sum()\
.reset_index()\
.sort_values("QUANTITYORDERED")\
.reset_index(drop=True)
Units_of_each_product.columns = ["Product Name", "Average price"]
Units_of_each_product.index += 1
Units_of_each_product
```

What does the total amount of merchandise sold?


```python
Total_Amount_of_merchandise_sold = df_1.groupby("PRODUCTLINE") ["SALES"].sum()\
.reset_index()
Total_Amount_of_merchandise_sold.columns = ["Product Name", "Sales"]
Total_Amount_of_merchandise_sold.index += 1
Total_Amount_of_merchandise_sold
```

How much money was generated in total sales of all products?

```python
Total_Sales = df_1['SALES'].sum()
Total_Sales
```

Which city generated the most revenue?a

```python
City_With_Highest_Revenue = df_1.groupby('CITY')['SALES'].sum().idxmax()
Highest_Revenue = df_1.groupby('CITY')['SALES'].sum().max()
print(f"The city with the highest revenue is {City_With_Highest_Revenue} with a revenue of {Highest_Revenue}")
```

Which city generated the least revenue?a

```python
City_With_Lowest_Revenue = df_1.groupby('CITY')['SALES'].sum().idxmin()
Lowest_Revenue = df_1.groupby('CITY')['SALES'].sum().min()
print(f"The city with the lowest revenue is {City_With_Lowest_Revenue} with a revenue of {Lowest_Revenue}")
```

Which city generated the most revenue?b

```python
Top_city = df_1.groupby('CITY')['SALES'].sum()\
.sort_values(ascending=False)\
.head(1)
Top_city = Top_city.reset_index()
Top_city.columns = ["Product Name", "Revenue"]
Top_city.index += 1
Top_city
```

Which city generated the least revenue?b

```python
Least_city = df_1.groupby('CITY')['SALES'].sum()\
.sort_values(ascending=True)\
.head(1)
Least_city = Least_city.reset_index()
Least_city.columns = ["Product Name", "Revenue"]
Least_city.index += 1
Least_city
```
